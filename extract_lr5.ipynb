```python
import os
import pandas as pd

pd.options.mode.chained_assignment = None
```

```python
def find_lr5(string):
    if 'lirads5' in string:
        print('contains lirads5')
    elif 'lr5' in string:
        print('contains lr5')
    elif 'liradscategory5' in string:
        print('contains liradscategory5')
    else:
        print('lirads5 NOT FOUND')
```

```python
dest = r'C:\Users\rob\Desktop\Work (LIG)\LIRADS_5_NLP\edited_spreadsheets'
spreadsheet = '2013 LI-RADS 5.csv'

df = pd.read_csv(spreadsheet)
df[['removed text', 'removed lirads legend text', 'trimmed text', 'contains lirads5?']] = ''
print(df.shape)
```

```python
for i in range(df.shape[0]):
    text = df['Report text'][i]
        
    if text.lower().find('**li-rads**') >= 0:
        text_no_hist = text[(text.lower().find('other')):]
        trimmed_text = text_no_hist[:((text_no_hist.lower().find('**li-rads**')))]

        removed_text = text[:(text.lower().find('other'))]
        removed_legend = text_no_hist[((text_no_hist.lower().find('**li-rads**'))):]
        
        if len(removed_legend) < 10:
            removed_legend = '0'
    
        df['trimmed text'][i] = trimmed_text
        df['removed text'][i] = removed_text
        df['removed lirads legend text'][i] = removed_legend
    
        new_string = ''.join(filter(str.isalnum, trimmed_text)).lower()
    
        if 'lirads5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        elif 'lr5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        elif 'liradscategory5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        else:
            df['contains lirads5?'][i] = str('lirads5 NOT FOUND')
            
    elif text.lower().find('**lirads**') >= 0:
        text_no_hist = text[(text.lower().find('other')):]
        trimmed_text = text_no_hist[:((text_no_hist.lower().find('**lirads**')))]

        removed_text = text[:(text.lower().find('other'))]
        removed_legend = text_no_hist[((text_no_hist.lower().find('**lirads**'))):]
        
        if len(removed_legend) < 10:
            removed_legend = '0'
    
        df['trimmed text'][i] = trimmed_text
        df['removed text'][i] = removed_text
        df['removed lirads legend text'][i] = removed_legend
    
        new_string = ''.join(filter(str.isalnum, trimmed_text)).lower()
    
        if 'lirads5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        elif 'lr5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        elif 'liradscategory5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        else:
            df['contains lirads5?'][i] = str('lirads5 NOT FOUND')
            
    else:
        text_no_legend = text[:((text.lower().find('**li-rads**')))]
        removed_legend = text[((text.lower().find('**li-rads**'))):]
        
        if len(removed_legend) < 10:
            removed_legend = '0'
        
        df['trimmed text'][i] = text_no_legend
        df['removed text'][i] = '0'
        df['removed lirads legend text'][i] = removed_legend
        
        new_string = ''.join(filter(str.isalnum, text_no_legend)).lower()
        
        if 'lirads5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        elif 'lr5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        elif 'liradscategory5' in new_string:
            df['contains lirads5?'][i] = str('yes')
        else:
            df['contains lirads5?'][i] = str('lirads5 NOT FOUND')
```
